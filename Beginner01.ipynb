```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

df = pd.read_csv('Salary dataset.csv')

#dataframe format in double brackets for input
sns.set(style="darkgrid")

X = df[['YearsExperience']] # Not a bug
y = df['Salary'] # Target

X_train, _, y_train, _ = train_test_split(X, y, train_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

palette = sns.color_palette("Blues", 6)
blue1 = palette[4]
blue2 = palette[2]

# plot the data

sns.regplot(data=df, x='YearsExperience', y='Salary', color=blue2, line_kws={"color": blue1}).set(title='Linear Regression: Years of Experience vs Salary')

#Display the plot
plt.show()
```
